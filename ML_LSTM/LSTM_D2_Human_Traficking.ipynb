```python
import pandas as pd

df = pd.read_csv("./data_glotip.csv")
df = df.drop_duplicates()
df = df.drop(columns=['Iso3_code', 'Region', 'Subregion', 'Dimension', 'Source', 'Indicator', 'Category' ])
drop_pop = df[df['Unit of measurement'] == 'Rate per 100,000 population'].index
df = df.drop(drop_pop)
```

```python
df['Country'] = df['Country'].str.strip()
print(df.isnull().any())
```

```python
df = df.dropna(subset=['Country'])
print(df.isnull().any())
```

```python
df
```

```python
drop_pop = df[df['Unit of measurement'] == 'Rate per 100,000 population'].index
df = df.drop(drop_pop)
df
```

```python
#value data type
#year data type
#gender categorization
```

```python
df['Year'] = pd.to_datetime(df.Year, format='%Y')
print(df.dtypes)
```

```python
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
unique_countries = df['Country'].unique()
print(unique_countries)
n = len(pd.unique(df['Country']))
print(n)
```

```python
lack_data = {country: len(data) for country, data in df.groupby('Country') if len(data) < 100}
lack_data
```

```python
countries_to_drop = list(lack_data.keys())

# Drop rows from the DataFrame for these countries
df = df[~df['Country'].isin(countries_to_drop)]
df
```

```python
df = df.drop(columns=['Unit of measurement' ])
#handle the Sex column
unique_genders = df['Sex'].unique()
unique_genders
```

```python
#making "other" and "unknown" values as "total" gender
df['Sex'] = df['Sex'].replace(["Other", "Unknown"], "Total")
print(df["Sex"].unique())
```

```python
# Generate dummy variables for 'Sex'
df = pd.get_dummies(df, columns=['Sex'], prefix='Gender')

# Convert dummy columns to integers
df = df.astype({'Gender_Female': 'int', 'Gender_Male': 'int', 'Gender_Total': 'int'})
df
```

```python
#handle age column
unique_age =  df['Age'].unique()
unique_age
```

```python
df['Age'] = df['Age'].replace(["Unknown"], "Total")
print(df["Age"].unique())
```

```python
# Generate dummy variables for 'Sex'
df = pd.get_dummies(df, columns=['Age'], prefix='Age')
print(df.columns)
```

```python
# Convert dummy columns to integers
df = df.astype({'Age_Total': 'int', 'Age_0 to 17 years': 'int', 'Age_18 years or over': 'int'})
```

```python
df = df.rename(columns={'Age_0 to 17 years': 'Age_Under18', 'Age_18 years or over': 'Age_Over18'})
df
```

```python
#handle txtVALUE
print(df["txtVALUE"].unique())
```

```python
non_numeric_rows = df[df['txtVALUE'].str.contains(r'[^0-9]', regex=True)]
non_numeric_rows = df[df['txtVALUE'].str.contains(r'[^0-9]', regex=True)]

# Get unique values
unique_non_numeric_values = non_numeric_rows['txtVALUE'].unique()

# Display all unique non-numeric values
print(unique_non_numeric_values)
```

```python
df['VALUE'] = df['txtVALUE'].str.replace(',', '', regex=True)
df['VALUE'] = df['VALUE'].replace({
    '<5': '4'
})
df['VALUE'] = pd.to_numeric(df['VALUE'], errors='coerce').astype('Int64')  # Nullable integer type

df
```

```python
df = df.drop(columns=['txtVALUE'])
df
```

```python
df.to_csv('D2_data.csv', index=False)
```
